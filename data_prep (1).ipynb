```python
for file in files:
    path = os.path.join(folder, file)
    new_path = os.path.join("combined", file)
    images = np.load(path)
    im_shape = images.shape
    im_height = im_shape[1]
    im_width = im_shape[2]
    canvas = np.zeros((im_height*18, im_width*20))
    height = 0
    batch = 0
    for i in range(18):
        width = 0
        canvas[height : height+im_height, :]
        for idx, image in enumerate(images[batch: batch+20]):
            canvas[height : height+im_height, width: width+im_width] = image
            width += im_width
        batch += 20
        height += im_height

    np.save(new_path, canvas)
    
        
```

```python
plt.imshow(canvas)
```
